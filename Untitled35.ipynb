**1- Uploading the Dataset**

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("/archive (1).zip")

# Display the first few rows
df.head()
```

**2-Understand the data structure and meta data**

```python
# Check dataset dimensions
print(f"Dataset shape: {df.shape}")

# Get column names and data types
print(df.info())

# Summary statistics for numerical features
print(df.describe())
```

**3-Handle Missing Data**

```python

# Convert 'date' column to datetime format (if needed)
df['date'] = pd.to_datetime(df['date'], errors='coerce')

# Identify numerical columns only
numeric_cols = df.select_dtypes(include=['number']).columns

# Fill missing values in numeric columns with their median
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())

# Display missing values count after handling
print(df.isnull().sum())
```

**4. Pairplot for Relationships**

```python
import seaborn as sns  # Import seaborn for visualization
import matplotlib.pyplot as plt  # Import matplotlib for plotting

# Create pairplot
sns.pairplot(df)
plt.show()
```

**5: Save the Cleaned Data**

```python
df.to_csv("cleaned_dataset.csv", index=False)
```
